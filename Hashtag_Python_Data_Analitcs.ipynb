```python
import pandas as pd
import numpy as np
```

```python
df_vendas = pd.read_csv('vendas_tech.csv', low_memory=False)
print(df_vendas)
df_gerentes = pd.read_excel('gerentes_lojas.xlsx')
print(df_gerentes)

```

```python
# inspecao dos dados

print(df_vendas)

# me mostre as primeiras 15 linhas do df
print(df_vendas.head(15))

# me mostre as ultimas 15 linhas do df
print(df_vendas.tail(15))

# me mostre 15 linhas de forma aleatoria do df
print(df_vendas.sample())

# me mostre quantas linhas e colunas o df tem, no formato 100000,8
print(df_vendas.shape())

# me mostre quais colunas existem no meu df
print(df_vendas.columns)

# me mostre um resumo do df. quais sao os tipos das colunas, sao str, float, int, memory usage, index, ...
print(df_vendas.info())

# me mostre a mean, std, quartis, max, min, count
print(df_vendas.describe())




```

```python
# tratamento de dados
print(df_vendas['Loja'])
print(df_vendas[['Loja', 'Cliente']])




# colunas

# exclua a coluna do df chamada de 'Data_Base' (ela so tem 1 linha preenchida, por isso vai ser excluida)
df_analise = df_vendas.drop(columns==['Data_Base'])




# nulos

#df_analise = df_analise.dropna() # excluir todas as linhas quem tem pelo menos 1 valor vazio

# Preenche os valores NaN com "Online". A função fillna() substitui os valores ausentes (NaN). Neste caso, o programador decidiu considerar que toda loja sem informação será identificada como "Online".

df_analise['Loja'] = df_analise['Loja'].fillna('Online')



# tipos de dados

# coloque o dtype da coluna de 'Data' do df do tipo de str que esta atualmente para datatime. e formate a data nesse estilo '%Y-%m-%d' = 2026-12-31
df_analise['Data'] = pd.to_datetime(df_analise['Data'],format='%Y-%m-%d')



# padronizacao da escrita ex (SP) (Sp) (sp)

# padronize a coluna 'Loja' do df. tirando os espacos em branco das laterias do texto ex( Sao Paulo ) = (Sao Paulo)
df_analise['Loja'] = df_analise['Loja'].str.strip()
# padronize a coluna 'Loja' do df. colocando a primeira palavra em maiusculo ex (sao paulo) or (SAO PAULO) = (Sao Paulo)
df_analise['Loja'] = df_analise['Loja'].str.title()



# duplicatas

# tire todas as linhas duplicadas, onde o id_pedido se repete
df_analise = df_analise.drop_duplicates(subset=['ID_Pedido'])


# # me mostre um novo resumo do df. quais sao os tipos das colunas, sao str, float, int, memory usage, index, ...
print(df_analise)
print(df_analise.info())

```

```python
# analise de dados



# criar novas colunas
# faturamento
df_analise['Faturamento'] = df_analise['Qtd'] * df_analise['Preco_Unitario']


# forma de venda

df_analise['Forma_de_Venda'] = np.where (df_analise['Loja'] == 'Online', 'Online', 'Presencial')



# regiao

print(df_analise['Loja'].unique())

dic_regioes = {'Sao Paulo':'Sudeste','Belo Horizonte':'Sudeste','Online':'Online','Rio de Janeiro':'Sudeste','Salvador':'Nordeste','Recife':'Nordeste','Curitiba':'Sul','Porto Alegre':'Sul'}

df_analise['Regiao'] = df_analise['Loja'].map(dic_regioes)

print(df_analise)
```
