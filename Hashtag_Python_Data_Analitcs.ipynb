```python
# inspecao dos dados

print(df_vendas)

# me mostre as primeiras 15 linhas do df
print(df_vendas.head(15))

# me mostre as ultimas 15 linhas do df
print(df_vendas.tail(15))

# me mostre 15 linhas de forma aleatoria do df
print(df_vendas.sample())

# me mostre quantas linhas e colunas o df tem, no formato 100000,8
print(df_vendas.shape())

# me mostre quais colunas existem no meu df
print(df_vendas.columns)

# me mostre um resumo do df. quais sao os tipos das colunas, sao str, float, int, memory usage, index, ...
print(df_vendas.info())

# me mostre a mean, std, quartis, max, min, count
print(df_vendas.describe())




```

```python
# tratamento de dados
print(df_vendas['Loja'])
print(df_vendas[['Loja', 'Cliente']])




# colunas

# exclua a coluna do df chamada de 'Data_Base' (ela so tem 1 linha preenchida, por isso vai ser excluida)
df_analise = df_vendas.drop(columns==['Data_Base'])




# nulos

#df_analise = df_analise.dropna() # excluir todas as linhas quem tem pelo menos 1 valor vazio

# Preenche os valores NaN com "Online". A função fillna() substitui os valores ausentes (NaN). Neste caso, o programador decidiu considerar que toda loja sem informação será identificada como "Online".

df_analise['Loja'] = df_analise['Loja'].fillna('Online')



# tipos de dados

# coloque o dtype da coluna de 'Data' do df do tipo de str que esta atualmente para datatime. e formate a data nesse estilo '%Y-%m-%d' = 2026-12-31
df_analise['Data'] = pd.to_datetime(df_analise['Data'],format='%Y-%m-%d')



# padronizacao da escrita ex (SP) (Sp) (sp)

# padronize a coluna 'Loja' do df. tirando os espacos em branco das laterias do texto ex( Sao Paulo ) = (Sao Paulo)
df_analise['Loja'] = df_analise['Loja'].str.strip()
# padronize a coluna 'Loja' do df. colocando a primeira palavra em maiusculo ex (sao paulo) or (SAO PAULO) = (Sao Paulo)
df_analise['Loja'] = df_analise['Loja'].str.title()



# duplicatas

# tire todas as linhas duplicadas, onde o id_pedido se repete
df_analise = df_analise.drop_duplicates(subset=['ID_Pedido'])


# # me mostre um novo resumo do df. quais sao os tipos das colunas, sao str, float, int, memory usage, index, ...
print(df_analise)
print(df_analise.info())

```

```python
# analise de dados



# criar novas colunas
# crie uma coluna no df. com base nas colunas do df.(qtd) & (preco_unitario) faca uma multiplicacao entre os 2, e coloque isso na variavel chamada de ('Faturamento')
df_analise['Faturamento'] = df_analise['Qtd'] * df_analise['Preco_Unitario']


# forma de venda
# crie uma coluna nova no df. com base em um filtro em cima de uma coluna ja exisitente no df. chamado de (Loja). para criar o filtro  o programador usa o where. dentro da coluna ('Loja') caso exista ('Online') escreva na mesma linha so que agora na nova coluna chamada de ('Forma_de_Pagamento') o nome ('Online'). caso exista ('Rio de Janeiro','Sao Paulo','Curitiba')... escreva na forma de pagamento o nome chamado de Presencial, para o programador entender se a forma de pagamento foi online ou presencial depedendo se a compra foi online ou presencial
df_analise['Forma_de_Venda'] = np.where (df_analise['Loja'] == 'Online', 'Online', 'Presencial')



# regiao

# me mostre a coluna ('Loja') do df. mas quero apenas os valores unicos, nesse caso: apenas as lojas unicas ex ('Rio de Janeiro','Sao Paulo','Curitiba')...
print(df_analise['Loja'].unique())

# aqui ele fez um dicionario, estilo o aurelio. quero pesquisar como se chama maça em english. vou no aurelio e busco oque corresponde a maça em english, e recebo como retorno apple. aqui e simplismente isso, observe que ele colocou o estado e a regiao dele, para quando o python procurar por sao paulo esse cod vai retornar para o python que a regiao e sudeste
dic_regioes = {'Sao Paulo':'Sudeste','Belo Horizonte':'Sudeste','Online':'Online','Rio de Janeiro':'Sudeste','Salvador':'Nordeste','Recife':'Nordeste','Curitiba':'Sul','Porto Alegre':'Sul'}

# estou criando uma coluna chamada de ('Regiao') no df. o python vai receber a coluna ('Loja') e com isso, o python vai requisitar o cod python (dic_regioes) que funciona como um aurelio para o python nesse caso, assim o programador ajudou a explicar o que o python precisa mapear do dicionario que o entreguei. e compara-lo com o df. e assim ele vai ver oque bate nas informacoes (aqui eu o entreguei ('Sao Paulo')) e na coluna ('Loja') do df tambem tem essa linha, entao o python vai entender oque vai colocar na nova coluna ('Regiao') com base no dicionario, ali eu mandei que para {'Sao Paulo':'Sudeste'} e oque vai aparecer na regiao da nova coluna e o sudeste como pedi ao python
df_analise['Regiao'] = df_analise['Loja'].map(dic_regioes)

# me mostre as atualizacoes do df. agora com a coluna regiao adicionada
print(df_analise)

# conte todos os nulos de todas as colunas do meu df. quero saber se apos essa adicao da coluna ('Regiao') o df ficou comrrompido
print(df_analise.isna().sum())
```
