```python
import pandas as pd
import numpy as np
```

```python
# df = pd.read_csv('')
```

```python
dados = {
    'Nome':['Ana','Bruno','Carlos','Daniela','Eduardo'],
    'idade':['25','30','35','40','46'],
    'Area':['Adm','Financeiro','Adm','Sales','Financeiro'],
    'Salario':['3000','5000','7000','9000','11000']
}


df = pd.DataFrame(dados)

display(df)
```

```python
#describe

df.describe()
```

```python
# aplicar aumento no salario dos funcionarios

df.info()

df['Salario'] = df['Salario'].astype(float)


df['Salario Ajustado'] = df['Salario'].apply(lambda x: x * 1.1)
display(df)
```

```python
# quero a media de salario por area


df_media_area = df.groupby('Area').mean(numeric_only=True)
display(df_media_area)
```

```python
# tire as linhas com valores NaN

mean_salarial = df['Salario'].mean()
df['Salario'] = df['Salario'].fillna(mean_salarial)
df['Salario Ajustado'] = df['Salario Ajustado'].fillna(df['Salario Ajustado'].mean)
display(df)
```
