```python
# Existem imóveis com muitos reviews e outros com poucos reviews?

# criei uma variavel chamada de mean_reviews para armazenar o resultado, apos isso puxei do dataframe a media do numero de reviews. apos isso mostrei para o prog (Gustavo Shelby)

mean_reviews = df['number_of_reviews'].mean()

display('A media do numero de Reviews eh:',mean_reviews)

# criei uma variavel chamada de median_reviews. puxei o df e fiz a mediana com o numero de reviews, apos isso mostrei na tela
median_reviews = df['number_of_reviews'].median()

display('A mediana do numero de Reviews eh:',median_reviews)


# com isso mostrei a media e mediana da quantidade de reviews por imovel, respondendo a pergunta. para mostrar a media se existem muitos ou poucos reviews por imovel










```
