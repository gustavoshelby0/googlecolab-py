```python
import pandas as pd
import numpy as np
```

```python
df_vendas = pd.read_csv('vendas_tech.csv', low_memory=False)
print(df_vendas)
df_gerentes = pd.read_excel('gerentes_lojas.xlsx')
print(df_gerentes)

```

```python
# inspecao dos dados

print(df_vendas)

# me mostre as primeiras 15 linhas do df
print(df_vendas.head(15))

# me mostre as ultimas 15 linhas do df
print(df_vendas.tail(15))

# me mostre 15 linhas de forma aleatoria do df
print(df_vendas.sample())

# me mostre quantas linhas e colunas o df tem, no formato 100000,8
print(df_vendas.shape())

# me mostre quais colunas existem no meu df
print(df_vendas.columns)

# me mostre um resumo do df. quais sao os tipos das colunas, sao str, float, int, memory usage, index, ...
print(df_vendas.info())

# me mostre a mean, std, quartis, max, min, count
print(df_vendas.describe())




```

```python
# tratamento de dados
print(df_vendas['Loja'])
print(df_vendas[['Loja', 'Cliente']])




# colunas

# exclua a coluna do df chamada de 'Data_Base' (ela so tem 1 linha preenchida, por isso vai ser excluida)
df_analise = df_vendas.drop(columns==['Data_Base'])




# nulos

#df_analise = df_analise.dropna() # excluir todas as linhas quem tem pelo menos 1 valor vazio

# Preenche os valores NaN com "Online". A função fillna() substitui os valores ausentes (NaN). Neste caso, o programador decidiu considerar que toda loja sem informação será identificada como "Online".

df_analise['Loja'] = df_analise['Loja'].fillna('Online')



# tipos de dados

# coloque o dtype da coluna de 'Data' do df do tipo de str que esta atualmente para datatime. e formate a data nesse estilo '%Y-%m-%d' = 2026-12-31
df_analise['Data'] = pd.to_datetime(df_analise['Data'],format='%Y-%m-%d')



# padronizacao da escrita ex (SP) (Sp) (sp)

# padronize a coluna 'Loja' do df. tirando os espacos em branco das laterias do texto ex( Sao Paulo ) = (Sao Paulo)
df_analise['Loja'] = df_analise['Loja'].str.strip()
# padronize a coluna 'Loja' do df. colocando a primeira palavra em maiusculo ex (sao paulo) or (SAO PAULO) = (Sao Paulo)
df_analise['Loja'] = df_analise['Loja'].str.title()

df_analise['Loja'] = df_gerentes['Loja'].str.strip()

df_analise['Loja'] = df_gerentes['Loja'].str.title()






# duplicatas

# tire todas as linhas duplicadas, onde o id_pedido se repete
df_analise = df_analise.drop_duplicates(subset=['ID_Pedido'])


# # me mostre um novo resumo do df. quais sao os tipos das colunas, sao str, float, int, memory usage, index, ...
print(df_analise)
print(df_analise.info())

```

```python
# analise de dados



# criar novas colunas
# crie uma coluna no df. com base nas colunas do df.(qtd) & (preco_unitario) faca uma multiplicacao entre os 2, e coloque isso na variavel chamada de ('Faturamento')
df_analise['Faturamento'] = df_analise['Qtd'] * df_analise['Preco_Unitario']


# forma de venda
# crie uma coluna nova no df. com base em um filtro em cima de uma coluna ja exisitente no df. chamado de (Loja). para criar o filtro  o programador usa o where. dentro da coluna ('Loja') caso exista ('Online') escreva na mesma linha so que agora na nova coluna chamada de ('Forma_de_Pagamento') o nome ('Online'). caso exista ('Rio de Janeiro','Sao Paulo','Curitiba')... escreva na forma de pagamento o nome chamado de Presencial, para o programador entender se a forma de pagamento foi online ou presencial depedendo se a compra foi online ou presencial
df_analise['Forma_de_Venda'] = np.where (df_analise['Loja'] == 'Online', 'Online', 'Presencial')



# regiao

# me mostre a coluna ('Loja') do df. mas quero apenas os valores unicos, nesse caso: apenas as lojas unicas ex ('Rio de Janeiro','Sao Paulo','Curitiba')...
print(df_analise['Loja'].unique())

# aqui ele fez um dicionario, estilo o aurelio. quero pesquisar como se chama maça em english. vou no aurelio e busco oque corresponde a maça em english, e recebo como retorno apple. aqui e simplismente isso, observe que ele colocou o estado e a regiao dele, para quando o python procurar por sao paulo esse cod vai retornar para o python que a regiao e sudeste
dic_regioes = {'Sao Paulo':'Sudeste','Belo Horizonte':'Sudeste','Online':'Online','Rio de Janeiro':'Sudeste','Salvador':'Nordeste','Recife':'Nordeste','Curitiba':'Sul','Porto Alegre':'Sul'}

# estou criando uma coluna chamada de ('Regiao') no df. o python vai receber a coluna ('Loja') e com isso, o python vai requisitar o cod python (dic_regioes) que funciona como um aurelio para o python nesse caso, assim o programador ajudou a explicar o que o python precisa mapear do dicionario que o entreguei. e compara-lo com o df. e assim ele vai ver oque bate nas informacoes (aqui eu o entreguei ('Sao Paulo')) e na coluna ('Loja') do df tambem tem essa linha, entao o python vai entender oque vai colocar na nova coluna ('Regiao') com base no dicionario, ali eu mandei que para {'Sao Paulo':'Sudeste'} e oque vai aparecer na regiao da nova coluna e o sudeste como pedi ao python
df_analise['Regiao'] = df_analise['Loja'].map(dic_regioes)

# me mostre as atualizacoes do df. agora com a coluna regiao adicionada
print(df_analise)

# conte todos os nulos de todas as colunas do meu df. quero saber se apos essa adicao da coluna ('Regiao') o df ficou comrrompido
print(df_analise.isna().sum())
```

```python
# analisar -> filtrar

df_analise = df_analise.short_values(by=['Data', 'Faturamento'])
df_analise = df_analise.reset_index(drop=True)




# .iloc: id do pedido -> por nome da coluna e por nome da linha]

id_pedido = 4
loja = df_analise.loc[df_analise['ID_Pedido']==4, 'Loja']
Produto = df_analise.loc[df_analise['ID_Pedido']==4, 'Produto']
Cliente = df_analise.loc[df_analise['ID_Pedido']==4, 'Cliente']
print('Loja','Produto','Cliente')





# .loc -> por posicao

id_pedido = 4
loja = df_analise.loc[3,2]
Produto = df_analise.loc[3,3]
Cliente = df_analise.loc[3,6]
print('Loja','Produto','Cliente')



# condicional
# df_id_pedido_4 = df_analise[df_analise['ID_Pedido']==4]




# exportar pedacos do df

df_vendas_sp = df_analise [df_analise['Loja']=='Sao Paulo']
df_vendas_sp.to_csv('Vendas_SP.csv', index=False)

print(df_vendas_sp)


# exportar as vendas de 2024




df_vendas_2024 = df_analise[df_analise['Data'].dt.year=='2024-01-01']

print(df_vendas_2024)





















# duplas condicoes



# condicao1 = df_analise['Produto']=='Cabo HDMI'
# condicao2 = df_analise['Regiao']=='Sul'
# df_vendas_hdmi_sul = df_analise[condicao1 & condicao2]



df_vendas_hdmi_sul = df_analise[(df_analise['Produto']=='Cabo HDMI') & (df_analise['Regiao']=='Sul')]

print(df_vendas_hdmi_sul)
```

```python
# analises por agrupamento
# print(df_analise)




# rank de faturamento por loja

# coloque na variavel (analise_lojas) um agrupamento de lojas e faturamento por loja: ou seja, quero que voce me fale a soma do faturamento por loja, e me mostre apenas as colunas de faturamento e loja no resultado
analise_lojas = df_analise[('Loja','Faturamento')].groupby('Loja').sum()

# coloque na variavel (analise_lojas) criada anteriormente, agora fazendo um ordenamento decrecente na coluna (faturamento) para saber quais lojas vendem mais, no estilo de rank (do maior para o menor)
analise_lojas = analise_lojas.short_values(by='Faturamento', asceding=False)

# reset o index do python que existe nesse df. apos o ordernamento anterior o index ficou bagurancado, essa funcao corriji isso, apos ter sido feito o ordenamento por faturamento
analise_lojas = analise_lojas.reset_index()


# faca uma subistituicao da coluna ("Faturamento") do df. nessa subistituicao voce deve usar a funcao .map para analisar as linhas dessa coluna dentro df, e subistituir a mascara do numero do faturamento, antes por exemplo 400000 para R$ 400.000
analise_lojas['Faturamento'] = analise_lojas['Faturamento'].map('R$(:,.2f)'.format)


# me mostre o df agora com as modificacoes
print(analise_lojas)







# rank produtos que + venderam Online


# na variavel (df_analise_online) que um filtro dentro da coluna de ('Loja') para eu ter de retorno apenas as vendas Online. atualmente esta assim ex('SP','RJ','CE','Online') com esse filtro vou ter o retorno apenas das vendas feitas pela Loja Online
df_analise_online = df_analise[df_analise['Loja']=='Online']

# na variavel ('analise_produtos_online') crie um agrupamento do df, usando as colunas [['Produto','Qtd']] pelo produto. ou seja: quero a quantidade de vendas por produto
analise_produtos_online = df_vendas_online[['Produto','Qtd']].groupby('Produto').sum()

# quero renomear a coluna qtd dentro do df. de("Qtd") para ("Vendas Totais")
analise_produtos_online = analise_produtos_online.rename(columns={'Qtd': 'Vendas Totais'})

# me mostre o df
print(analise_produtos_online)







# analise de ranking por loja e por produto
# quais produtos venderam mais em cada uma das lojas


# quero a loja que mais vendeu, quero o produto que mais vendeu e a quantidade de vendas dele. ( [['','','']] aqui estou selecionando oque vou mostrar na tela, estilo um select do SQL) agrupando por produto e loja que gerou a venda
# me mostre o resultado da loja do produto e a quantidade de vendas por loja e produto. faca a soma da coluna quantidade

df_produtos_em_lojas = df_analise[['Loja','Produto','Qtd']].groupby(['Loja','Produto']).sum()

# faca um ordenamento de qual produto mais vendeu dentro do df, o resultado vai mostrar. qual loja e qual produto mais venderam de forma quantitativa
df_produtos_em_lojas = df_produtos_em_lojas.sort_values(by='Qtd', asceding=False)


# me mostre o df. com o agrupamento mostrando qual loja e qual produto vendeu mais
print(df_produtos_em_lojas)

```

```python
# quais gerentes bateram a meta em jan de 2023


# faca um filtro e adicione na variavel df_meta. quero dentro do df_analise apenas as vendas de 2023 de janeiro
df_meta = df_analise[(df_analise['Data']==2023 & df_analise['Data'].dt.month==1)]
# me mostre o df.
print(df_meta)


# me mostre a soma do faturamento por loja, em janeiro de 2023
# o as_index=False serve para nao deixar a Loja se tornar um index, mas continue sendo uma coluna. que possa sofrer analises posteriores tipo merge()
df_meta = df_meta[['Loja', 'Faturamento']].groupby('Loja', as_index=False).sum()

# faca um estilo procv or left join entre as tabelas (df_meta) & (df_gerentes)

df_meta = df_meta.merge(df_gerentes, on='Loja',how='left')
# faca um if or um where do mesmo tipo do SQL. se o faturamento da loja e maior ou igual a meta mensal, ela bateu a meta!. caso contrario ele nao bateu a meta
df_meta['Bateu Meta'] = np.where(df_meta['Faturamento'] >= df_meta['Meta_Mensal'],'Sim','Nao')
```
